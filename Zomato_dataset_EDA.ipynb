```python
df_country = pd.read_excel('Country-Code.xlsx')
```

```python
final_df = pd.merge(df,df_country,on='Country Code',how='left')
# on represent based on which column the data to be merged.
```

```python
final_df.head()
```

```python
# To check datatypes
final_df.dtypes
```

```python
final_df.Country.value_counts()
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_val = final_df.Country.value_counts().values
```

```python
country_name
```

```python
#pie-chart taking top three values
plt.pie(country_val[:3],labels=country_name[:3],autopct='%1.2f%%')
```

# Observation : 
    
    1. Zomato has strong market in india than anyother countries.
    2. Zomato maximum record or transcation are based on india after that the transcation are done in united kingdom and united states with stating value of 94.39,0.87 and 4.73.

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0 : 'Rating count'})
```

```python
ratings
```

# Observation : 
    
    1. The rating between 4.5 to 4.9 -- Excellent
    2. The rating between 4.0 to 4.4 -- Very good 
    3. The rating between 3.5 to 3.9 -- Good 
    4. The rating between 2.5 to 3.4 -- Average 
    5. The rating between 1.8 to 2.4 -- Poor

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green'])
```

# Observation : 
    1. Not rated count is very high.
    2. Maximum number of rating are between 2.5 to 3.4

```python
sns.countplot(x='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','darkgreen'])
```
